```python
import os
from openai import OpenAI
import re
client = OpenAI(api_key="")

messages = [

        {"role": "user", "content": "in the cash flow statement, there is a 'Cash Paid for Plant and Investments', can you explain it, When it is positive, when it is negative"}
    ]
response = client.chat.completions.create(
        model="gpt-4o",
        messages=messages,
        temperature=0
    )
response_message = response.choices[0].message.content
print(response_message )

```
